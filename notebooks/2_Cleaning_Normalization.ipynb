```python
import pandas as pd
from datasets import load_dataset

dataset = load_dataset("7Xan7der7/us_airline_sentiment")

df = pd.DataFrame(dataset["train"])
```

```python
# Keeping only the columns we need

df = df[['text', 'airline_sentiment']]

df.head()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
import re

def clean_tweet(text):
    text = text.lower()
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'https?://\S+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    
    return text
```

```python

sample = "@VirginAmerica I HATE this flight!!! http://badflight.com #fail"
print(f"Original: {sample}")
print(f"Cleaned:  {clean_tweet(sample)}")
```

```python
# Apply the cleaning function on our data

df['cleaned_text'] = df['text'].apply(clean_tweet)

df[['text', 'cleaned_text']].head()
```

```python
df.to_csv('../data/cleaned_tweets.csv', index=False)
```
